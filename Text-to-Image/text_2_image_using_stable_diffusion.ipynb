```python
import gradio as gr
from PIL import Image
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline

prompt = [
    "a pikachu is running wild in the snow with his horse",
    "a robot is helping a dog in the lake"
]

modelid = "CompVis/stable-diffusion-v1-4"
device = "cuda"
pipe = StableDiffusionPipeline.from_pretrained(
    modelid,
    revision="fp16",
    torch_dtype=torch.float16,
    use_auth_token="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"
)
pipe.to(device)

def generate(prompt, negative_prompt="", strength=8.5):
    prompt = [prompt, negative_prompt]
    with autocast(device):
        image = pipe(prompt, guidance_scale=strength)["sample"][0]
    image.save('generatedimage.png')
    return image

inputs = [
    gr.inputs.Textbox(label="Enter Prompt Text"),
    gr.inputs.Textbox(label="Enter Negative Prompt Text"),
    gr.inputs.Slider(label="Guidance Strength", minimum=0.1, maximum=10.0, default=8.5, step=0.1)
]
outputs = gr.outputs.Image(type="pil", label="Generated Image")

gr.Interface(
    fn=generate,
    inputs=inputs,
    outputs=outputs,
    title="TEXT-2-IMAGE",
    description="Generate images based on your prompt text.",
    examples=[["A beautiful sunset over the mountains."]]
).launch()


```
